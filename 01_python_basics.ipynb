```python
age = 45
print(f"年龄: {age}, 类型: {type(age)}")
height = 175.5
print(f"身高: {height} cm, 类型: {type(height)}")
patient_name = "张三"
print(f"患者姓名: {patient_name}, 类型: {type(patient_name)}")
is_diabetic = True
print(f"是否糖尿病: {is_diabetic}, 类型: {type(is_diabetic)}")
```
